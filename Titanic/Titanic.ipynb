```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import mean_squared_error, mean_absolute_error, mean_absolute_percentage_error, r2_score
```

```python
train=pd.read_csv('/content/Titanic/Data/train.csv')
train=train.dropna()
features=['Sex','Age','Fare','Pclass']
Y=train['Survived']
X=train[features]
X=pd.get_dummies(X)
train
```

```python
X_train, X_test, Y_train, Y_test=train_test_split(X,Y, test_size=0.15, random_state=0)
Y_train
```

```python
X_train
```

```python
Y_test=Y_test.reset_index(drop=True)
```

```python
sk_Model=RandomForestClassifier(n_estimators=100, max_depth=5, random_state=1)
```

```python
sk_Model.fit(X,Y)
```

```python
Y_pred=sk_Model.predict(X_test)
```

```python
Comparision=pd.DataFrame(Y_pred, columns=['Survived_Predicted'])
```

```python
Comparision['Survived_trained']=Y_test
```

```python
Comparision
```

```python
print(f"Mean squared error is: {mean_squared_error(Y_test, Y_pred)}")
print(f"Root mean squared error is: {np.sqrt(mean_squared_error(Y_test, Y_pred))}")
print(f"R squared error is: {r2_score(Y_test, Y_pred)}")
print(f"Mean absolute error is: {mean_absolute_error(Y_test, Y_pred)}")
print(f"Mean absolute percentage error is: {mean_absolute_percentage_error(Y_test, Y_pred)*100}")
```
